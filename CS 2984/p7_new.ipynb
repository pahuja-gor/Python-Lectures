```python
# Retrieving Data
import json
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

with open('graduates.json') as f:
    data_dict = json.load(f)
```

```python
# Data Cleaning and Preprocessing
data = pd.json_normalize(data_dict)
data.head()
```

```python
data.info()
```

```python
grouped_data = data.groupby('Education.Major').mean().copy()
grouped_data.head()
```

```python
grouped_data.info()
```

```python
grouped_data.index
```

```python
grouped_data.info()
```

```python
def stem_major(major):
    
    stem = ['Atmospheric Sciences and Meteorology', 'Biochemistry and Biophysics',
       'Biological Sciences', 'Chemical Engineering', 'Chemistry',
       'Civil Engineering', 'Computer Science and Math', 'Economics', 'Electrical Engineering',
        'Genetics, Animal and Plant', 'Geology', 'Information Services and Systems',
            'Mechanical Engineering', 'OTHER Geological Sciences',
       'OTHER Physical and Related Sciences', 'Operations Research', 'Other Engineering',
       'Pharmacology, Human and Animal', 'Physics and Astronomy', 
       'Plant Sciences', 'Statistics', 'Zoology, General']
    
    if major in stem:
        return 1
    return 0
```

```python
grouped_data = grouped_data.reset_index()
```

```python
grouped_data.head()
```

```python
grouped_data['STEM'] = grouped_data['Education.Major'].apply(stem_major).astype('category')
```

```python
grouped_data[['Education.Major','STEM']]
```

```python
%matplotlib inline
grouped_data['FemRatio'] = grouped_data['Demographics.Gender.Females'] / grouped_data['Demographics.Total']
grouped_data.plot.scatter(x='FemRatio',y='Salaries.Mean',c='STEM', colormap='winter')
```

# Unsupervised learning: kmeans clustering

```python
# Analyzing Data 1
from sklearn.cluster import KMeans

data_numerical = grouped_data.select_dtypes('number')
model = KMeans(n_clusters=4, random_state=0)
features = data_numerical.columns.to_list()
features.remove('Salaries.Mean')
features.remove('Salaries.Median')
features.remove('Salaries.Standard Deviation')
features.remove('Salaries.Highest')
features.remove('Salaries.Quantity')

model.fit(data_numerical[features])
grouped_data['Labels']= model.labels_
grouped_data['Labels'] = grouped_data['Labels'].astype('category')
```

```python
grouped_data.plot.scatter(x='FemRatio',y='Salaries.Mean',c='Labels', colormap='viridis')
```

# Supervised Learning: decision tree classifier

```python
train = grouped_data.sample(int(len(grouped_data)*.8), random_state=13).copy()
indexes_used_in_train = train.index
test = grouped_data.loc[~grouped_data.index.isin(indexes_used_in_train)].copy()
```

```python
features = grouped_data.columns.tolist()
features.remove('Education.Major')
features.remove('Year')
features.remove('STEM')
features
```

```python
from sklearn.tree import DecisionTreeClassifier

target = 'STEM'
model = DecisionTreeClassifier()

model.fit(train[features],train[target])
test['Predicted_STEM'] = model.predict(test[features])
```

```python
test[['Education.Major','STEM', 'Predicted_STEM']]
```

```python
sum(test['STEM']==test['Predicted_STEM']) /len(test)*100
```

# Supervised Learning: linear regression
## Correlation heatmap
A tool used to identify correlated columns for feature selection 

```python
grouped_data.corr().abs()['Salaries.Mean'].sort_values(ascending=False).head(20)
```

```python
features = grouped_data.corr().abs()['Salaries.Mean'].sort_values(ascending=False).head(20).index.to_list()
features.remove('Salaries.Mean')
features.remove('Salaries.Median')
features.remove('Salaries.Standard Deviation')
features.remove('Salaries.Highest')
features.remove('Salaries.Quantity')
features
```

```python
target = 'Salaries.Mean'
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(train[features], train[target])
predictions = model.predict(test[features])

rmse = mean_squared_error(predictions, test[target])**.5
rmse
```
